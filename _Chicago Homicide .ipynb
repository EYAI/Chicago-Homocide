# data Cleaning

```python
data.drop(['Record ID', 'Agency Code','Agency Name', 'Agency Type','Record Source','Relationship'], axis=1,inplace=True)
```

```python
# delete the unsolved crimes
data=data[data["Crime Solved"]=="Yes"]
data.drop(["Crime Solved"], axis=1,inplace=True)
```

mapping data

```python
data['Weapon']=data['Weapon'].map({'Blunt Object':0, 'Strangulation':1, 'Unknown':-1, 'Rifle':3, 'Knife':4, 'Firearm':5, 'Shotgun':6, 'Fall':7, 'Handgun':8, 'Drowning':9, 'Suffocation':10, 'Explosives':11, 'Fire':12,'Drugs':13, 'Gun':14, 'Poison':15})
data['Perpetrator Age']=pd.to_numeric(data['Perpetrator Age'], errors='coerce')
data['Perpetrator Sex'] = data['Perpetrator Sex'].map( {'Female': 0, 'Unknown': -1,'Male': 1} ).astype(int)
data['Victim Sex'] = data['Victim Sex'].map( {'Female': 0, 'Unknown': -1,'Male': 1} ).astype(int)
data['Month'] = data['Month'].map( {'January': 1, 'February': 2,'March': 3, 'April':4,'May':5,'June':6,'July':7,"August":8,"September":9,"October":10,"November":11,"December":12} ).astype(int)
data['Victim Ethnicity'] = data['Victim Ethnicity'].map({'Unknown':-1,'Not Hispanic':0,'Hispanic':1})
data['Perpetrator Ethnicity'] = data['Perpetrator Ethnicity'].map({'Unknown':-1,'Not Hispanic':0,'Hispanic':1})
data['Perpetrator Race'] = data['Perpetrator Race'].map({'Unknown':-1,'Native American/Alaska Native':1,'White':2, 'Black':3, 'Asian/Pacific Islander':4})
data['Victim Race'] = data['Victim Race'].map({'Unknown':-1,'Native American/Alaska Native':1,'White':2, 'Black':3, 'Asian/Pacific Islander':4})
#data['Relationship']=data['Relationship'].map({'Acquaintance':0, 'Unknown':-1, 'Wife':2, 'Stranger':3, 'Girlfriend':4, 'Ex-Husband':5, 'Brother':6, 'Stepdaughter':7, 'Husband':8, 'Sister':9, 'Friend':10, 'Family':11, 'Neighbor' :12,'Father' :13,'In-Law':14, 'Son':15, 'Ex-Wife':16, 'Boyfriend':18, 'Mother':17 ,'Common-Law Husband':19, 'Common-Law Wife':20, 'Stepfather':21, 'Stepson':22 ,'Stepmother':23 ,'Daughter':24 ,'Boyfriend/Girlfriend':25 ,'Employer' :26,'Employee':27})
data['Crime Type']=data['Crime Type'].map({'Murder or Manslaughter':2, 'Manslaughter by Negligence':1})
```

# Train Model

# 1 - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
Y=data[['Perpetrator Sex', 'Perpetrator Age', 'Perpetrator Race','Perpetrator Ethnicity']]
X=data[['Year', 'Month', 'Incident', 'Crime Type','Victim Sex', 'Victim Age', 'Victim Race', 'Victim Ethnicity',
        'Weapon', 'Victim Count']]
```

```python
# Perpetrator Sex
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30)
```

```python
logmodelsex = LogisticRegression()
logmodelsex.fit(X_train,y_train['Perpetrator Sex'])
```

```python
#logmodelage = LogisticRegression()
#logmodelage.fit(X_train,y_train['Perpetrator Age'])
```

```python
logmodelrace = LogisticRegression()
logmodelrace.fit(X_train,y_train['Perpetrator Race'])
```

```python
logmodelEthnicity = LogisticRegression()
logmodelEthnicity.fit(X_train,y_train['Perpetrator Ethnicity'])
```

# Predictions and Evaluations
Now predict values for the testing data.

```python
predictionsSex = logmodelsex.predict(X_test)
#predictionsAge = logmodelage.predict(X_test)
predictionsRace = logmodelrace.predict(X_test)
predictionsEthnicity = logmodelEthnicity.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test['Perpetrator Sex'],predictionsSex))
```

```python
#print(classification_report(y_test['Perpetrator Age'],predictionsAge))
```

```python
 print(classification_report(y_test['Perpetrator Race'],predictionsRace))
```

```python
print(classification_report(y_test['Perpetrator Ethnicity'],predictionsEthnicity))
```
